```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
injuries_data= pd.read_csv('KSI_CLEAN.csv',parse_dates=[['YEAR', 'MONTH','DAY']])
injuries_data['year'] = injuries_data['YEAR_MONTH_DAY'].dt.year
injuries_data['month'] = injuries_data['YEAR_MONTH_DAY'].dt.month
injuries_data['day'] = injuries_data['YEAR_MONTH_DAY'].dt.day
injuries_data_clean = injuries_data.replace(' ', np.nan, regex=False)
injuries_data_clean['WEEKDAY']= injuries_data_clean['WEEKDAY'].astype('str')
```

```python
# Columns with the percentage of Nan values
print(injuries_data_clean.isna().sum()/len(injuries_data_clean)*100)
```

```python
# Dropping the columns with more than 80% Nan values
injuries_data_clean = injuries_data_clean.drop(['OFFSET', 'PEDTYPE', 'PEDACT', 'PEDCOND', 'CYCLISTYPE', 'CYCACT', 'CYCCOND'], axis=1)
```

```python
# Changing the days od the week from numbers to string
for i in '0':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Monday'))
for i in '1':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Tuesday'))    
for i in '2':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Wednesday')) 
for i in '3':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Thursday'))
for i in '4':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Friday'))
for i in '5':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Saturday'))
for i in '6':
    injuries_data_clean['WEEKDAY'] = injuries_data_clean['WEEKDAY'].apply(lambda x: x.replace(i, 'Sunday'))

injuries_data_clean['WEEKDAY'].unique()
```

```python
#Number of Unique accidents by Year
Num_accident = injuries_data_clean.groupby('year')['ACCNUM'].nunique()
plt.figure(figsize=(12,6))
plt.title("Accidents in different years")
plt.ylabel('Number of Accidents (ACCNUM)')

ax = plt.gca()
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='red')
my_colors = list('rgbkymc')   #red, green, blue, black, etc.
Num_accident.plot(
    kind='bar', 
    color='blue',
    edgecolor='black'
)
plt.xticks(rotation=0)
#Num_accident.plot(kind='bar', color=my_colors)
plt.show()
```

```python
#Number of Unique accidents by Month
Num_accident = injuries_data_clean.groupby('month')['ACCNUM'].nunique()
plt.figure(figsize=(12,6))
plt.title("Accidents caused in different months")
plt.ylabel('Number of Accidents (ACCNUM)')

ax = plt.gca()
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='red')
my_colors = list('rgbkymc')   #red, green, blue, black, etc.
Num_accident.plot(
    kind='bar', 
    color='blue',
    edgecolor='black',
)
plt.xticks(rotation=0)
#Num_accident.plot(kind='bar',color= my_colors)
plt.show()
```

```python
# Changing all Non-Fatal injuries and Property Damage Only entries to Non-Fatal

List=['Non-Fatal Injury','Property Damage Only']
for i in List:
    injuries_data_clean['ACCLASS'] = injuries_data_clean['ACCLASS'].apply(lambda x: x.replace(i, 'Non-Fatal'))
    
injuries_data_clean['ACCLASS'].unique()
```

```python
#Categorizing Fatal vs. non-Fatal Incident
sns.catplot(x='year', kind='count', data=injuries_data_clean,  hue='ACCLASS')
plt.show()
```

```python
sns.catplot(x='year', kind='count', data=injuries_data_clean,  hue='FATAL')
```

```python
# 2013 has the highest number of Fatal accidents
Fatality = injuries_data_clean[injuries_data_clean['ACCLASS'] =='Fatal']
Fatality = Fatality.groupby(injuries_data_clean['year']).count()
plt.figure(figsize=(12,6))

plt.ylabel('Number of Injuries=Fatal')
Fatality['ACCLASS'].plot(kind='bar',color="blue" , edgecolor='black')
plt.grid()

plt.show()
```

```python
#Looking at area where accident happens

Region_injuries_data = injuries_data_clean['District'].value_counts()
plt.figure(figsize=(12,6))
plt.ylabel('Number of Accidents')
Region_injuries_data.plot(kind='bar',color=list('rgbkmc'))
plt.xticks(rotation=0)
plt.show()
```

```python
#Looking at Neighbourhood where accident happens

Hood_injuries_data = injuries_data_clean['Hood_Name'].value_counts()
plt.figure(figsize=(12,6))
plt.ylabel('Number of Accidents')
Hood_injuries_data.nlargest(10).plot(kind='bar')
plt.show()
```

```python
Weekday_injuries_data = injuries_data_clean['WEEKDAY'].value_counts()
plt.figure(figsize=(12,6))
plt.ylabel('Number of Accidents')
Weekday_injuries_data.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()
```

```python
vehicle_data=injuries_data_clean.pivot_table(index='year',margins=False,values=['CYCLIST', 'TRSN_CITY_VEH', 'MOTORCYCLE', 'TRUCK', 'EMERG_VEH'],aggfunc=np.sum)

vehicle_data.plot(figsize=(10,8), title="Type of vehicles involved in the accidents per year", grid=True)
plt.ylabel('Number of accidents')
plt.show()
```
